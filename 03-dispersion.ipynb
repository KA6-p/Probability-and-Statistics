## Generating a LOT of numbers

```python
num_samples = 10000
```

```python
uniform_vals = np.random.uniform(low=0.0,high=10.0,size=num_samples)
```

```python
sns.displot(uniform_vals,bins=20,kde=False)
plt.ylabel("Frequency")
plt.xlabel("Values")
plt.title("Uniform Values")
sns.despine(offset=10,trim=True);
plt.plot()
```

```python
normal_vals = np.random.normal(loc=5.0,scale=10.0,size=num_samples)
# More values closer t0 "loc" .... Sort of like [1 5 5 5 5 7]
```

```python
sns.displot(normal_vals,bins=20,kde=False)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Normal Values")
sns.despine(offset=10,trim=False);
plt.plot()
```

```python
print("Uniform vals mean:", np.mean(uniform_vals))
print("Normal vals mean:",np.mean(normal_vals))
```

```python
print("Uniform vals variance:",np.var(uniform_vals))
print("Normal vals variance:",np.var(normal_vals))
```

```python
print("Uniform vals std:",np.std(uniform_vals))
print("Normal vals std:",np.std(normal_vals))
```

```python
normal_vals = np.random.normal(loc=5.0,scale=10.0,size=num_samples)
nv_mean = np.mean(normal_vals)
nv_std = np.std(normal_vals)
```

```python
sns.displot(normal_vals,bins=20,kde=False)
plt.xlim(-60,60)

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Normal Values")
sns.despine(offset=10,trim=False);

# plot the SD line
x_c,y_c = ([nv_mean,nv_mean],[0,1500])
plt.plot(x_c,y_c,color='green',linewidth=2)

x_c,y_c = ([nv_mean + nv_std,nv_mean + nv_std],[0,1500])
plt.plot(x_c,y_c,color='red',linewidth=2)

x_c,y_c = ([nv_mean - nv_std,nv_mean - nv_std],[0,1500])
plt.plot(x_c,y_c,color='red',linewidth=2)

plt.show()

```

```python
normal_vals = np.random.normal(loc=5.0,scale=20.0,size=num_samples)
nv_mean = np.mean(normal_vals)
nv_std = np.std(normal_vals)
```

```python
sns.displot(normal_vals,bins=20,kde=False)
plt.xlim(-60,60)

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Normal Values")
sns.despine(offset=10,trim=False);

# plot the SD line
x_c,y_c = ([nv_mean,nv_mean],[0,1500])
plt.plot(x_c,y_c,color='green',linewidth=2)

x_c,y_c = ([nv_mean + nv_std,nv_mean + nv_std],[0,1500])
plt.plot(x_c,y_c,color='red',linewidth=2)

x_c,y_c = ([nv_mean - nv_std,nv_mean - nv_std],[0,1500])
plt.plot(x_c,y_c,color='red',linewidth=2)

plt.show()
```

```python
normal_vals_1 = np.random.normal(loc=5.0,scale=10.0,size=num_samples)
normal_vals_2 = np.random.normal(loc=5.0,scale=20.0,size=num_samples)

def plot_dis(vals,label):
    nv_mean = np.mean(vals)
    nv_std = np.std(vals)
    sns.displot(normal_vals,bins=20,kde=False)
    plt.xlim(-60,60)

    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.title(label)
    sns.despine(offset=10,trim=False);

    # plot the SD line
    x_c,y_c = ([nv_mean,nv_mean],[0,1500])
    plt.plot(x_c,y_c,color='green',linewidth=2)

    x_c,y_c = ([nv_mean + nv_std,nv_mean + nv_std],[0,1500])
    plt.plot(x_c,y_c,color='red',linewidth=2)

    x_c,y_c = ([nv_mean - nv_std,nv_mean - nv_std],[0,1500])
    plt.plot(x_c,y_c,color='red',linewidth=2)

    plt.show()

plot_dis(normal_vals_1,"Less Variant")
plot_dis(normal_vals_2,"More Variant")
```

```python
med_incs = [ x[0] for x in houses.data ]
mean = np.mean(med_incs)
std = np.std(med_incs)

sns.displot(med_incs,bins=20,kde=False)

  #plt.xlim(-60,60)

plt.xlabel("Median Income")
plt.ylabel("Frequency")
plt.title("Histogram of Median Incomes")
sns.despine(offset=10,trim=False);

 # plot the SD line
x_c,y_c = ([mean,mean],[0,4000])
plt.plot(x_c,y_c,color='green',linewidth=2)

x_c,y_c = ([mean + std,mean + std],[0,4000])
plt.plot(x_c,y_c,color='red',linewidth=2)

x_c,y_c = ([mean - std,mean - std],[0,4000])
plt.plot(x_c,y_c,color='red',linewidth=2)

plt.show()


```
