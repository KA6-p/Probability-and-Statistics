## Data Visualization

```python
y=[6,7,2,4,5]
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(y)
plt.show()
```

```python
plt.figure(figsize=(10,5))
# add a line renderer with legend and line thickness
plt.plot(y)
# to make figure readable
plt.title("First Example Plot")
plt.xlabel("x vals (s)")
plt.ylabel("y vals (m)")
plt.grid()
plt.show()
```

```python
# plotting weights
plt.figure(figsize=(20,5))
# add a line renderer with legend and line thickness
x = range(da['BMXWT'].size)
y = da['BMXWT']

plt.scatter(x,y) #marker='x',s=0.1

# to make figure readable
plt.title("Weights of Individuals")
plt.xlabel("ID of Individuals")
plt.ylabel("Weight (kg)")
plt.grid()
plt.show()
```

```python
# plotting weights
plt.figure(figsize=(20,5))
# add a line renderer with legend and line thickness
x = range(da['BMXWT'].size)
y = da['BMXWT']

plt.scatter(x,y,marker='x',s=0.2)

# to make figure readable
plt.title("Weights of Individuals")
plt.xlabel("ID of Individuals")
plt.ylabel("Weight (kg)")

plt.show()
```
